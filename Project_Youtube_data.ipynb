# 5. Convert datatype of column published_at from object to pandas datetime.

```python
data['published_at']=pd.to_datetime(data['published_at'])
data.dtypes
```

# 6. Create a new column as 'published_month' using the column published_at (display the months only)


```python
data['published_month']=data['published_at'].dt.month
data.head(5)
```

#  7.Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and so on.....


```python
month={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
data['published_month']=data['published_month'].replace(month)
data.head(5)
```

# 8.Find the number of videos published each month and arrange the months in a decreasing order based on the video count

```python
video_count=data.groupby(data['published_month'])['video_id'].count()
video_count=video_count.sort_values(ascending=False)
print(video_count)
```

# 9. Find the count of unique video_id, channel_id and channel_title.


```python
video_count=data['video_id'].nunique()
channel_id_count=data['channel_id'].nunique()
channel_title_count=data['channel_title'].nunique()
print("Number of unique video_id values:",video_count)
print("Number of unique channel_id values:",channel_id_count)
print("Number of unique channel_title values:",channel_title_count)
```

# 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos

```python
video_count_by_channel=data.groupby('channel_title')['video_id'].count()
top10=video_count_by_channel.nlargest(10)
bottom10=video_count_by_channel.nsmallest(10)
print("Top 10 channel names with highest number of videos:",top10)
print("\nBottom 10 channel names with lowest number of videos:",bottom10)
```

# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


- Number of views effect on how many people disliked the video.
- As we can in the scatterplot that as less number of views for a video has less dislikes 
- We can also see that as the number of views for videos increases the dislikes for that video is also increasing
- We can also see that number of views and people disliked coloumns are positively correlated 
- By this we can say that the number of views has effect on how many people disliked the video

```python
plt.scatter(data['view_count'],data['dislikes'])
plt.title('Views vs. Dislikes')
plt.xlabel('Number of Views')
plt.ylabel('Number of Dislikes')
plt.show()
correlation=data['view_count'].corr(data['dislikes'])
print("Correlation coefficient between views and dislikes:",correlation)
```
